# Importing the Packages

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading dataset

```
df=pd.read_csv("/content/drive/My Drive/The Sparks Foundation Tasks/Datasets/task_3_and_task_4_dataset.csv")
df.head()
```

# Performing Label encoding

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["Species"] = le.fit_transform(df["Species"])
df.head()
```

```
x=df.iloc[:,[1,2,3,4]]
y = df.iloc[:,5]
```

# Performing the visualization of the dataset

```
x1=df["SepalLengthCm"]
y1=df["Species"]
with sns.axes_style('white'):
    sns.jointplot(x1, y1, df, kind='kde')
```

```
x1=df["SepalWidthCm"]
y1=df["Species"]
sns.swarmplot(y1, x1)
```

```
sns.barplot(x="PetalLengthCm",y="Species",data=df,estimator=np.mean)
```

```
sns.catplot(x="PetalWidthCm",y="Species",data=df)
```

# Applying Decision Tree Classifier

```
from sklearn import tree
model= tree.DecisionTreeClassifier()
```

```
from sklearn.model_selection import train_test_split
(xtrain,xtext,ytrain,ytest)= train_test_split(x,y,test_size = 0.2, random_state =0)
```

```
model.fit(xtrain,ytrain)
model.score(xtrain,ytrain)
```

```
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```
dtree= DecisionTreeClassifier().fit(x, y)
plt.figure(figsize = (20,10))
plot_tree(dtree, rounded = True, filled=True)
plt.show()
```
